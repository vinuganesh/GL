### Find out the IQR for the column - Age 

*Hint: IQR - Interquartile range = 75th percentile - 25th percentile* 

```python
hr['Age'].describe()
```

```python
# 1st Q
hr['Age'].quantile(.25)
```

```python
# 3rd Q
hr['Age'].quantile(.75)
```

```python
# median
hr['Age'].quantile(.5)
```

```python
#IGR
IGR=hr['Age'].quantile(.75)-hr['Age'].quantile(.25)
IGR
```
